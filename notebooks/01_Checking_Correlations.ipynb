### 1. Correlations of psy. features (N=5) with average PSD features (N=580)

```python
import pandas as pd 
import numpy as np

df = pd.read_csv("data/df_meg_psy_allfeatures.csv")

df.head(2)
```

```python
df.isnull().sum().sum()
```

```python
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

```python
df.shape
```

```python
vector_bdi = df['BDI']
vector_moca = df['MoCA corretto']
vector_updrs3 = df['UPDRS III ']
vector_mhy = df['mH&Y']
vector_apathy = df['Apathy']
```

```python
feature_matrix = df.iloc[:, 19:]
```

```python
print(feature_matrix.shape)
print(vector_bdi.shape)
```

```python
corrs = pd.DataFrame(dtype='float64', columns=['spearman_corr', 'p_val1', 'sign1', 
                                               'pearson_corr', 'p_val2', 'sign2'])
```

```python
corrs
```

```python
from scipy.stats import spearmanr, pearsonr

for i in range(feature_matrix.shape[1]):
    for vec in [vector_bdi, vector_moca, vector_updrs3, vector_mhy, vector_apathy]:
        # Spearman Correlation
        corr, p_val = spearmanr(vec, feature_matrix.iloc[:, i]) 
        corrs.loc[f'{vec.name}_{feature_matrix.columns[i]}', ['spearman_corr', 'p_val1', 'sign1']] = corr, \
                                        p_val, 'negative' if corr < 0 else 'positive'
        # Pearson Correlation
        corr, p_val = pearsonr(vec, feature_matrix.iloc[:, i])
        corrs.loc[f'{vec.name}_{feature_matrix.columns[i]}', ['pearson_corr', 'p_val2', 'sign2']] = corr, \
                                        p_val, 'negative' if corr < 0 else 'positive'
```

```python
corrs
```

```python
corrs_spearman = corrs[['spearman_corr', 'p_val1']]
corrs_spearman['abs_corr'] = corrs_spearman['spearman_corr'].abs()
corrs_spearman = corrs_spearman.sort_values(by='abs_corr', ascending=False).drop(columns=['abs_corr'])
corrs_spearman
```

```python
corrs_pearson = corrs[['pearson_corr', 'p_val2']]
corrs_pearson['abs_corr'] = corrs_pearson['pearson_corr'].abs()
corrs_pearson = corrs_pearson.sort_values(by='abs_corr', ascending=False).drop(columns=['abs_corr'])
corrs_pearson
```

```python
corrs_psd = corrs_spearman.merge(corrs_pearson, left_index=True, right_index=True, how='inner')
```

```python
corrs_psd
```

```python
corrs_psd.to_csv("../output/correlations_psd_features_all.csv", index=True)
corrs_pearson.to_csv("../output/correlations_psd_features_pearson.csv", index=True)
corrs_spearman.to_csv("../output/correlations_psd_features_spearman.csv", index=True)
```

```python
pd.read_csv("../output/correlations_psd_features_pearson.csv", index_col=0)
```
